```python
#background=np.expand_dims(np.array([0.16,0.34,0.33,0.17]),axis=1)
background=np.expand_dims(np.array([0.25]*4),axis=1)
```

## ATAC ## 

```python
shap_atac_center_6mer=np.loadtxt("deepshap.pwm.deepshap.atac.gm12878.0.computed.6mer.center.profile")
shap_atac_sum_6mer=np.loadtxt("deepshap.pwm.deepshap.atac.gm12878.0.computed.6mer.sum.profile")
shap_atac_first_6mer=np.loadtxt("deepshap.pwm.deepshap.atac.gm12878.0.computed.6mer.first.profile")
f,axes=plt.subplots(1,3,dpi=80,figsize=(16,4))
matplotlib.pyplot.subplots_adjust(wspace=0.4,hspace=0.4)
axes[0]=plot_bases_on_ax(shap_atac_sum_6mer.T,axes[0],show_ticks=True)
axes[0].set_title("6-mer, sum, GM12878 DeepSHAP Profile atac")
axes[1]=plot_bases_on_ax(shap_atac_first_6mer.T,axes[1],show_ticks=True)
axes[1].set_title("6-mer, first, GM12878 DeepSHAP Profile atac")
axes[2]=plot_bases_on_ax(shap_atac_center_6mer.T,axes[2],show_ticks=True)
axes[2].set_title("6-mer, center, GM12878 DeepSHAP Profile atac")
```

```python
#load pfm
atac_sum_6mer=np.loadtxt("atac.gm12878.0.computed.6mer.sum")
atac_sum_24mer=np.loadtxt("atac.gm12878.0.computed.24mer.sum")
atac_first_6mer=np.loadtxt("atac.gm12878.0.computed.6mer.first")
atac_first_24mer=np.loadtxt("atac.gm12878.0.computed.24mer.first")
atac_center_6mer=np.loadtxt("atac.gm12878.0.computed.6mer.center")
atac_center_24mer=np.loadtxt("atac.gm12878.0.computed.24mer.center")
#get pssm from pfm
atac_sum_6mer_pssm=np.log2(atac_sum_6mer/background)
atac_sum_24mer_pssm=np.log2(atac_sum_24mer/background)
atac_first_6mer_pssm=np.log2(atac_first_6mer/background)
atac_first_24mer_pssm=np.log2(atac_first_24mer/background)
atac_center_6mer_pssm=np.log2(atac_center_6mer/background)
atac_center_24mer_pssm=np.log2(atac_center_24mer/background)
```

```python
f,axes=plt.subplots(4,3,dpi=80,figsize=(12,10))
matplotlib.pyplot.subplots_adjust(wspace=0.4,hspace=0.4)

axes[0,0]=plot_bases_on_ax(atac_sum_6mer.T,axes[0,0],show_ticks=True)
axes[0,0].set_title("6-mer, sum, PFM")
axes[1,0]=plot_bases_on_ax(atac_sum_6mer_pssm.T,axes[1,0],show_ticks=True)
axes[1,0].set_title("6-mer, sum, PSSM")

axes[2,0]=plot_bases_on_ax(atac_sum_24mer.T,axes[2,0],show_ticks=True)
axes[2,0].set_title("24-mer, sum, PFM")
axes[3,0]=plot_bases_on_ax(atac_sum_24mer_pssm.T,axes[3,0],show_ticks=True)
axes[3,0].set_title("24-mer, sum, PSSM")

axes[0,1]=plot_bases_on_ax(atac_first_6mer.T,axes[0,1],show_ticks=True)
axes[0,1].set_title("6-mer, first, PFM")
axes[1,1]=plot_bases_on_ax(atac_first_6mer_pssm.T,axes[1,1],show_ticks=True)
axes[1,1].set_title("6-mer, first, PSSM")

axes[2,1]=plot_bases_on_ax(atac_first_24mer.T,axes[2,1],show_ticks=True)
axes[2,1].set_title("24-mer, first, PFM")
axes[3,1]=plot_bases_on_ax(atac_first_24mer_pssm.T,axes[3,1],show_ticks=True)
axes[3,1].set_title("24-mer, first, PSSM")



axes[0,2]=plot_bases_on_ax(atac_center_6mer.T,axes[0,2],show_ticks=True)
axes[0,2].set_title("6-mer, center, PFM")
axes[1,2]=plot_bases_on_ax(atac_center_6mer_pssm.T,axes[1,2],show_ticks=True)
axes[1,2].set_title("6-mer, center, PSSM")

axes[2,2]=plot_bases_on_ax(atac_center_24mer.T,axes[2,2],show_ticks=True)
axes[2,2].set_title("24-mer, center, PFM")
axes[3,2]=plot_bases_on_ax(atac_center_24mer_pssm.T,axes[3,2],show_ticks=True)
axes[3,2].set_title("24-mer, center, PSSM")

plt.suptitle("ATAC, GM12878, fold 0")
```
