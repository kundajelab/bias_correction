```python
%matplotlib inline 
import matplotlib 
from matplotlib import pyplot as plt
from kerasAC.vis.plot_letters import * 
import numpy as np 
```

## ATAC ## 

```python
#load pfm
atac_sum_6mer=np.loadtxt("pwm.atac.gm12878.0.computed.6mer.sum")
atac_sum_24mer=np.loadtxt("pwm.atac.gm12878.0.computed.24mer.sum.new")
atac_first_6mer=np.loadtxt("pwm.atac.gm12878.0.computed.6mer.first")
atac_first_24mer=np.loadtxt("pwm.atac.gm12878.0.computed.24mer.first.new")
atac_center_6mer=np.loadtxt("pwm.atac.gm12878.0.computed.6mer.center")
atac_center_24mer=np.loadtxt("pwm.atac.gm12878.0.computed.24mer.center.new")
```

```python
f,axes=plt.subplots(2,3,dpi=80,figsize=(14,10))
matplotlib.pyplot.subplots_adjust(wspace=0.4,hspace=0.4)

axes[0,0]=plot_bases_on_ax(atac_sum_6mer.T,axes[0,0],show_ticks=True)
axes[0,0].set_title("BPNET ATAC, 6-mer, sum, PSSM")

axes[1,0]=plot_bases_on_ax(atac_sum_24mer.T,axes[1,0],show_ticks=True)
axes[1,0].set_title("BPNET ATAC, 24-mer, sum, PSSM")

axes[0,1]=plot_bases_on_ax(atac_first_6mer.T,axes[0,1],show_ticks=True)
axes[0,1].set_title("BPNET ATAC, 6-mer, first, PSSM")


axes[1,1]=plot_bases_on_ax(atac_first_24mer.T,axes[1,1],show_ticks=True)
axes[1,1].set_title("BPNET ATAC, 24-mer, first, PSSM")


axes[0,2]=plot_bases_on_ax(atac_center_6mer.T,axes[0,2],show_ticks=True)
axes[0,2].set_title("BPNET ATAC, 6-mer, center, PSSM")

axes[1,2]=plot_bases_on_ax(atac_center_24mer.T,axes[1,2],show_ticks=True)
axes[1,2].set_title("BPENT ATAC, 24-mer, center, PSSM")

plt.suptitle("ATAC, GM12878, fold 0")
plt.show()
```
